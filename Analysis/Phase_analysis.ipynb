```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft
A = 0.5
freq = 10
phase = 30
f_s = 32*freq

t = np.linspace(0, 2, f_s*2)
phi = phase * np.pi/180
x = A * np.cos(2 * np.pi * freq * t + phi)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(t, x, label='cosine function')
plt.title('Waveform Plot')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude (V)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
freqs = fft.fftshift(fft.fftfreq(x.size, 1 / f_s))
X = fft.fft(x)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(freqs, X)
plt.title('Frequency spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Frequency Domain Magnitude')
plt.xlim(-f_s , f_s )
plt.grid(True)
plt.show()
```

```python
phase_angle = np.arctan2(np.imag(X), np.real(X)) * 180 / np.pi

plt.figure(figsize=(20, 10))
plt.plot(freqs, phase_angle)
plt.title('Frequency Phase Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Frequency Domain Phase [grad]')
plt.xlim(-f_s , f_s)
plt.grid(True)
plt.show()
```
